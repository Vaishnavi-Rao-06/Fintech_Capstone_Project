```python
return_cols = ['return_1yr_pct','return_3yr_pct', 'return_5yr_pct', 'benchmark_3yr_pct']
```

```python
for col in return_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')
    print(f"{col} - nulls after conversion: {data[col].isnull().sum()}")
```

```python
data['sharpe_ratio'] = pd.to_numeric(data['sharpe_ratio'], errors='coerce')
```

```python
negative_sharpe = data[data['sharpe_ratio']<0]
print(f"Funds with negative sharpe ratio:{len(negative_sharpe)}")
```

```python
data['negative_Sharpe_flag'] = data['sharpe_ratio']<0
data[['scheme_name','sharpe_ratio','negative_Sharpe_flag']].head(10)
```

```python
data['expense_ratio_pct'] = pd.to_numeric(data['expense_ratio_pct'], errors='coerce')
```

```python
out_of_range = data[(data['expense_ratio_pct'] <0.1) | (data['expense_ratio_pct']>2.5)]
print(f"out of range expense ratios : {len(out_of_range)}")
```

```python
if len(out_of_range)>0:
    print(out_of_range[['scheme_name','expense_ratio_pct']])
else:
    print("All expense ratios within 0.1% - 2.5% range.")
```
