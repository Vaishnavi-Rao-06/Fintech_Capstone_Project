## 📊 Key EDA Findings

### Finding 1 — NAV Growth
HDFC Short Term Debt Fund showed the highest NAV growth from ₹3,200 to ₹4,500+ between 2022–2026, significantly outperforming peers. *(Ref: Chart 1 — NAV Trends)*


### Finding 2 — AUM Dominance
SBI Mutual Fund dominates with ₹12.5L Crore AUM, nearly double the next competitor, reflecting investor trust in PSU-backed funds. *(Ref: Chart 2 — AUM Growth)*

### Finding 3 — SIP All-Time High
Monthly SIP inflows hit an all-time high of ₹31,002 Crore in December 2025, reflecting growing retail investor participation. *(Ref: Chart 3 — SIP Inflows)*

### Finding 4 — Liquid Fund Dominance
Liquid funds consistently show the highest net inflows across all months, acting as parking vehicles for short-term investor money. *(Ref: Chart 4 — Category Heatmap)*

### Finding 5 — Young Investors Leading
The 25–35 age group dominates investor distribution, indicating strong millennial participation in mutual fund investments. *(Ref: Chart 5 — Age Distribution)*

### Finding 6 — T30 Cities Dominate
T30 cities account for over 70% of total SIP investments, with Maharashtra and Delhi leading in absolute transaction volumes. *(Ref: Chart 6 — Geographic Distribution)*

### Finding 7 — Folio Count Doubled
Total folios nearly doubled from 13.26 Crore in January 2022 to 26.12 Crore in December 2025, reflecting massive mutual fund adoption. *(Ref: Chart 7 — Folio Growth)*

### Finding 8 — High Correlation Among Equity Funds
Top 10 equity funds show high positive correlation (>0.85), suggesting they move together during market rallies and corrections. *(Ref: Chart 8 — Correlation Matrix)*

### Finding 9 — Financial Services Dominates
Financial services sector accounts for the largest portfolio allocation across equity funds, followed by IT and Consumer Goods. *(Ref: Chart 9 — Sector Allocation)*

### Finding 10 — Gender Gap in Investing
Male investors account for nearly 65% of transactions, highlighting a significant gender gap in mutual fund participation in India. *(Ref: Chart 5c — Gender Split)*
