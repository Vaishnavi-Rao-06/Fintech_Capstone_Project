```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///../data/bluestock_mf.db')
print("Database engine created!")
```

```python
data = pd.read_csv("../data/processed/clean_fund_master.csv")
data.to_sql('dim_fund', engine, if_exists='replace', index=False)
print(f"dim_fund loaded — {len(data)} rows")
```

```python
data = pd.read_csv("../data/processed/clean_nav.csv")
data.to_sql('fact_nav', engine, if_exists='replace', index=False)
print(f"fact_nav loaded — {len(data)} rows")
```

```python
data = pd.read_csv("../data/processed/clean_transactions.csv")
data.to_sql('fact_transactions', engine, if_exists='replace', index=False)
print(f"fact_transactions loaded — {len(data)} rows")
```

```python
data = pd.read_csv("../data/processed/clean_performance.csv")
data.to_sql('fact_performance', engine, if_exists='replace', index=False)
print(f"fact_performance loaded — {len(data)} rows")
```

```python
tables = {
    'clean_aum': 'aum_by_fund_house',
    'clean_sip_inflows': 'sip_inflows',
    'clean_category_inflows': 'category_inflows',
    'clean_folio_count': 'folio_count',
    'clean_portfolio_holdings': 'portfolio_holdings',
    'clean_benchmark_indices': 'benchmark_indices'
}
```

```python
for file, table in tables.items():
    data = pd.read_csv(f"../data/processed/{file}.csv")
    data.to_sql(table, engine, if_exists='replace', index=False)
    print(f"{table} loaded — {len(data)} rows")
```

```python
from sqlalchemy import inspect
```

```python
inspector = inspect(engine)
tables = inspector.get_table_names()
```

```python
print(f"Total tables in database: {len(tables)}")
for table in tables:
    print(f"  {table}")
```
