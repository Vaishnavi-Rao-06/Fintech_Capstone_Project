```python
data['date'] = pd.to_datetime(data['date'], format='mixed', dayfirst=True)
```

```python
print(data.dtypes)
```

```python
data.head()
```

```python
data = data.sort_values(['amfi_code', 'date']).reset_index(drop=True)
```

```python
data.head(10)
```

```python
before = len(data)
data = data.drop_duplicates(subset=['amfi_code','date'])
after = len(data)
```

```python
print(f"duplicates removed :{before - after}")
```

```python
print(f"Remaining rows :{after}")
```

```python
data = data.set_index('date')
data = data.groupby('amfi_code')['nav'].apply(
    lambda x: x.resample('B').asfreq().ffill()
)
data = data.reset_index(level=0).reset_index()
data.columns = ['date', 'amfi_code', 'nav']

print(data.shape)
data.head()
```

```python
invalid = data[data['nav'] <= 0]
print(f"Invalid nav rows (<=0): {len(invalid)}")
if len(invalid) >0:
    print(invalid)
else:
    print("Valid")
```

```python
data.to_csv("C:/Users/vaishnavi/Downloads/Fintech_Capstone_Project/data/processed/clean_nav.csv",index=False)
```
