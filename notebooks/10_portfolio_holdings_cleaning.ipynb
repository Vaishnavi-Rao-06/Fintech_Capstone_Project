```python
invalid_weight = data[(data['weight_pct'] <= 0) | (data['weight_pct'] > 100)]
print(f"Invalid weight rows: {len(invalid_weight)}")
```

```python
if len(invalid_weight) == 0:
    print("valid")
```

```python
invalid_value = data[(data['market_value_cr'] <= 0) | (data['current_price_inr'] <= 0)]
print(f"Invalid market value/price rows: {len(invalid_value)}")
```

```python
if len(invalid_value) == 0:
    print("valid")
```

```python
weight_sum = data.groupby('amfi_code')['weight_pct'].sum()
```
