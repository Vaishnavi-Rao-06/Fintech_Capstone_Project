```python
folio_cols = ['total_folios_crore', 'equity_folios_crore', 'debt_folios_crore', 'hybrid_folios_crore',
              'others_folios_crore']
```

```python
for col in folio_cols:
    invalid = data[data[col] <= 0]
    print(f"{col} invalid rows: {len(invalid)}")
```

```python
data['calculated_total'] = (data['equity_folios_crore'] + data['debt_folios_crore'] + 
                          data['hybrid_folios_crore'] + data['others_folios_crore'])
```

```python
mismatch = data[round(data['calculated_total'], 2) != round(data['total_folios_crore'], 2)]
```

```python
print(f"Mismatch rows: {len(mismatch)}")
```

```python
if len(mismatch) == 0:
    print("Total folios match sum of categories!")
```

```python
print(mismatch[['month', 'total_folios_crore', 'calculated_total']])
mismatch['difference'] = mismatch['total_folios_crore'] - mismatch['calculated_total']
print(mismatch['difference'].describe())
```

```python
data['folio_rounding_flag'] = abs(data['total_folios_crore'] - data['calculated_total']) > 0.01
print(f"Actual mismatches (>0.01): {data['folio_rounding_flag'].sum()}")
print("All differences are just rounding errors — data is clean!")
```
