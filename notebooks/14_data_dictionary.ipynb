```python
import pandas as pd
```

```python
datasets = {
    'clean_fund_master': '01_fund_master.csv',
    'clean_nav': '02_nav_history.csv',
    'clean_aum': '03_aum_by_fund_house.csv',
    'clean_sip_inflows': '04_monthly_sip_inflows.csv',
    'clean_category_inflows': '05_category_inflows.csv',
    'clean_folio_count': '06_industry_folio_count.csv',
    'clean_performance': '07_scheme_performance.csv',
    'clean_transactions': '08_investor_transactions.csv',
    'clean_portfolio_holdings': '09_portfolio_holdings.csv',
    'clean_benchmark_indices': '10_benchmark_indices.csv'
}
```

```python
output = "# Data Dictionary\n\n"

for cleaned_file, raw_file in datasets.items():
    data = pd.read_csv(f"../data/processed/{cleaned_file}.csv")
    
    output += f"## {cleaned_file}\n"
    output += f"**Source:** `data/raw/{raw_file}`  \n"
    output += f"**Rows:** {data.shape[0]} | **Columns:** {data.shape[1]}\n\n"
    output += "| Column | Type | Description |\n"
    output += "|--------|------|-------------|\n"
    
    for col in data.columns:
        dtype = str(data[col].dtype)
        output += f"| {col} | {dtype} | - |\n"
    
    output += "\n---\n\n"
```

```python
with open("../reports/data_dictionary.md", "w") as f:
    f.write(output)
```
