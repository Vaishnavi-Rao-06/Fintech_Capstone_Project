```python
import pandas as pd
```

```python
fund_master = pd.read_csv("C:/Users/vaishnavi/Downloads/Fintech_Capstone_Project/data/raw/01_fund_master.csv")
nav_history = pd.read_csv("C:/Users/vaishnavi/Downloads/Fintech_Capstone_Project/data/raw/02_nav_history.csv")
```

```python
print("Fund master shape:", fund_master.shape)
print("Nav history shape:", nav_history.shape)
```

```python
print("Fund master columns:", fund_master.columns.tolist())
print("Nav history columns:", nav_history.columns.tolist())
```

```python
fund_codes = set(fund_master['amfi_code'])
nav_codes = set(nav_history['amfi_code'])                
```

```python
matched = fund_codes & nav_codes
missing = fund_codes - nav_codes
extra = nav_codes - fund_codes
```

```python
print(f"codes in fund_master:{len(fund_codes)}")
print(f"codes in nav_history:{len(nav_codes)}")
print(f"matched codes :{len(matched)}")
print(f"missing codes :{len(missing)}")
print(f"extra codes :{len(extra)}")
```

```python
merged = pd.merge(fund_master, nav_history, on='amfi_code', how='left',indicator = True)
```

```python
print(merged['_merge'].value_counts())
```

```python
report = f"""
========== DATA QUALITY REPORT ==========
Fund Master   : {len(fund_codes)} unique AMFI codes
NAV History   : {len(nav_codes)} unique AMFI codes

Matched     : {len(matched)}
Missing     : {len(missing)} (in fund_master but NOT in nav_history)
Extra       : {len(extra)} (in nav_history but NOT in fund_master)

NAV History null check:
{nav_history.isnull().sum().to_string()}

Fund Master null check:
{fund_master.isnull().sum().to_string()}
==========================================
"""
```

```python
print(report)
```

```python
with open("../reports/data_quality_report.txt", "w") as f:
    f.write(report)

print("Report saved to reports/data_quality_report.txt ")
```
